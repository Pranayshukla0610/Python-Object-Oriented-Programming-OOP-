```python
#SPLITWISE SYSTEM

from abc import ABC, abstractmethod

class User:
  def __init__(self,name):
    self.name = name
    self.balance = {}

  def update_balance(self,other,amount):
    self.balance[other] = self.balance.get(other,0) + amount

class Expense(ABC):
  @abstractmethod
  def split(self):
    pass

class EqualExpense(Expense):
  def __init__(self,amount,users):
    self.users = users
    self.amount = amount

  def split(self):
    share = self.amount / len(self.users)
    return {user:share for user in self.users}

class Splitwise:
  def __init__(self):
    self.users = []

  def add_user(self,user):
    self.users.append(user)

  def add_expense(self,expense,player):
    splits = expense.split()

    for users, amount in splits.items():
      if user != player:
        user.update_balance(player.name, -amount)
        player.update_balance(player.name, amount)
```

```python
#ONLINE AUCTION SYSTEM

class Bid:
  def __init__(self,user,amount):
    self.user = user
    self.amount = amount

class Auction:
  def __init__(self):
    self.bids = []

  def place_bid(self,bid):
    self.bids.append(bid)

  def highest_bid(self):
    return max(self.bids, key = lambda bid:bid.amount)
```
