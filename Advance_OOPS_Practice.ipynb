```python
#ATM MACHINE SYSTEM

class Account:
  def __init__(self,owner,balance=0):
    self.owner = owner
    self.__balance = balance

  def deposit(self,amount):
    self.__balance += amount

  def withdraw(self,amount):
    if amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Not Enough Balance")

  def get_balance(self):
    return self.__balance


class ATM:
  def __init__(self):
    self.accounts = {}

  def create_account(self,owner):
    self.accounts[owner] = Account(owner)

  def get_account(self,owner):
    return self.accounts.get(owner)


atm = ATM()
atm.create_account('Alice')

acc = atm.get_account('Alice')
acc.deposit(500)
acc.withdraw(200)
```

```python
print(acc.get_balance())
```

```python
#ELEVATOR SYSTEM
class Elevator:
  def __init__(self):
    self.current_floor = 0

  def move_to(self,floor):
    print("Moving from", self.current_floor, "to", floor)
    self.current_floor = floor

class Building:
  def __init__(self):
    self.elevator = Elevator()

  def request_elevator(self,floor):
    self.elevator.move_to(floor)

b = Building()
b.request_elevator(5)
```

```python
#Music Playlist System

class Song:
  def __init__(self,title):
    self.title = title

class Playlist:
  def __init__(self):
    self.songs = []

  def add_song(self,song):
    self.songs.append(song)

  def play(self):
    for s in self.songs:
      print("Playing",s.title)

p = Playlist()
p.add_song(Song("Mujhe Tumse Pyaar Hua"))
p.add_song(Song("Saaki Saaki"))
p.play()
```

```python
#Chat Messaging System

class Message:
  def __init__(self,sender,receiver,text):
    self.sender = sender
    self.receiver = receiver
    self.text = text

class ChatRoom:
  def __init__(self):
    self.messages = []

  def send_message(self, message):
    self.messages.append(message)

  def show_message(self):
    for m in self.messages:
      print(m.sender,":",m.text)

room = ChatRoom()
room.send_message(Message("Alice","Bob","Hello"))
room.send_message(Message("Bob","Alice","How are you"))

room.show_message()
```

```python
#Task Scheduler System

class Task:
  def __init__(self,name):
    self.name = name
    self.completed = False

  def complete(self):
    self.completed = True

class Scheduler:
  def __init__(self):
    self.task = []

  def add_task(self,task):
    self.task.append(task)

  def show_task (self):
    for t in self.task:
      status = "Done" if t.completed else "Pending"
      print(t.name,status)

s = Scheduler()
t = Task("Study DSA in Python")
s.add_task(t)
t.complete()
s.show_task()
```
