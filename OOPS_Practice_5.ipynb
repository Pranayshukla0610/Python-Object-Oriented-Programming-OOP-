```python
#ONLINE VOTING SYSTEM

class Candidates:
  def __init__(self,name):
    self.name = name
    self.votes = 0

class VotingSystem:
  def vote(self,candidate):
    candidate.votes += 1
```

```python
#FILE UPLOAD SYSTEM

class File:
  def __init__(self,name):
    self.name = name

class Storage:
  def __init__(self):
    self.files = []

  def upload(self,file):
    self.files.append(file)
```

```python
#Multiplayer Game Lobby

class Player:
  def __init__(self,name):
    self.name = name

class Lobby:
  def __init__(self):
    self.players = []

  def join(self, player):
    self.players.append(player)
```

```python
#ONLINE LEARNING PLATFORM

class Course:
  def __init__(self,name):
    self.name = name

class Student:
  def __init__(self):
    self.courses = []

  def enroll(self,course):
    self.courses.append(course)
```

```python
#JOB PORTAL SYSTEM

class Job:
  def __init__(self,role):
    self.role = role

class JobPortal:
  def __init__(self):
    self.jobs = []

  def post_jon(self,job):
    self.jobs.append(job)
```

```python
#EMAIL SYSTEM

class Email:
  def __init__(self,sender,msg):
    self.sender = sender
    self.msg = msg

class EmailSystem:
  def __init__(self):
    self.emails = []

  def send_email(self,email):
    self.emails.append(email)
```
