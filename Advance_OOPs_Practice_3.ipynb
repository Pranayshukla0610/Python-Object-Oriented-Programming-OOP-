```python
#Online Auction System

class Bid:
  def __init__(self,user,amount):
    self.user = user
    self.amount = amount

class Auction:
  def __init__(self):
    self.highest_bid = None

  def place_bid(self,bid):
    if not self.highest_bid or bid.amount> self.highest_bid.amount:
      self.highest_bid = bid

  def winner(self):
    if self.highest_bid:
      print("Winner:",self.highest_bid.user)

a = Auction()
a.place_bid(Bid("Alice",100))
a.place_bid(Bid("Bob",200))

a.winner()
```

```python
#Task Scheduler System

class Task:
  def __init__(self,name):
    self.name = name
    self.completed = None

  def complete(self):
    self.complete = True


class Scheduler:
  def __init__(self):
    self.tasks = []

  def add_task(self,task):
    self.tasks.append(task)

  def show_tasks(self):
    for t in self.tasks:
      status = "Completed" if t.completed else "Pending"
      print(f"{t.name} - {status}")

s = Scheduler()
t1 = Task("Study OOP's")
s.add_task(t1)

t1.complete()

s.show_tasks()
```

```python
#Ride Sharing Matching System

class Driver:
  def __init__(self,name):
    self.name = name
    self.available = True

class Rider:
  def __init__(self,name):
    self.name = name

class Rideservice:
  def __init__(self):
    self.drivers = []

  def add_driver(self,driver):
    self.drivers.append(driver)

  def request_ride(self,rider):
    for d in self.drivers:
      if d.available:
        d.available = False
        print(rider.name, "matched with", d.name)
        return
    print("No drivers available")

service = Rideservice()
service.add_driver(Driver("Alice"))
service.add_driver(Driver("Bob"))

service.request_ride(Rider("Charlie"))
```

```python
#Chat messaging system

class Message:
  def __init__(self,sender,content):
    self.sender = sender
    self.content = content

class ChatRoom:
  def __init__(self):
    self.messages = []

  def send_messages(self,message):
    self.messages.append(message)

  def display_message(self):
    for m in self.messages:
      print(m.sender,":",m.content)

room = ChatRoom()
room.send_messages(Message("Alice","Hello,How are you?"))
room.send_messages(Message("Bob","I'm good,thanks!"))

room.display_message()
```

```python
#Music Playlist System

class Music:
  def __init__(self,name,singer):
    self.name = name
    self.singer = singer

class PlaySong:
  def __init__(self):
    self.songs = []

  def add_song(self,song):
    self.songs.append(song)

  def play_song(self):
    for s in self.songs:
      print("Playing",s.name, "by", s.singer)


p = PlaySong()
p.add_song(Music("layi vi na gayi","Sukhwinder Singh"))
p.add_song(Music("Tum Tak","Javed Ali"))
p.play_song()
```

```python
#Parking Garage System

class Car:
  def __init__(self,name,plate):
    self.name = name
    self.plate = plate

class ParkingGarage:
  def __init__(self,capacity):
    self.capacity = capacity
    self.cars = []

  def add_car(self,car):
    if len(self.cars) < self.capacity:
      self.cars.append(car)
      print("Car parked successfully")
    else:
      print("Parking garage is full")

  def leave(self,plate):
    for c in self.cars:
      if c.plate == plate:
        self.cars.remove(c)
        print("Car left")

garage = ParkingGarage(2)
garage.add_car(Car("Hyundai Creta","A1"))
garage.add_car(Car("Mahindra Thar","B1"))
garage.leave("B3")
```
