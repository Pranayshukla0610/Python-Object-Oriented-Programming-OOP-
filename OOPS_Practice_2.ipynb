```python
#UBER LIKE SYSTEM

class Driver:
  def __init__(self,name):
    self.name = name
    self.available = True

class Rider:
  def __init__(self,name):
    self.name = name

class Riderservice:
  def __init__(self):
    self.drivers = []

  def add_driver(self,driver):
    self.drivers.append(driver)

  def request_ride(self,rider):
    for d in self.drivers:
      if d.available:
        d.available = False
        print(f"{rider.name} is riding with {d.name}")
        return

    print("No drivers are availability")
```

```python
#Operator Overloading
class Number:
  def __init__(self,value):
    self.value = value

  def __add__(self,other):
    return Number(self.value + other.value)

n1 = Number(10)
n2 = Number(20)
n3 = n1 + n2
print(n3.value)
```

```python
#Method Overriding

class Parent:
  def show(self):
    print("Parent")

class Child:
  def show(self):
    print("Child")

c = Child()
c.show()
```
